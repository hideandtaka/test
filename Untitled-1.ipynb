```python
# 今日の売り上げを計算する簡単な売上計算機

def calculate_daily_sales(sales):
    """
    sales は [(商品名, 数量, 単価), …] のリストとする
    例: [("りんご", 10, 120), ("バナナ", 5, 80)]
    戻り値は売上の合計金額（整数または浮動小数点）。
    """
    return sum(qty * price for _, qty, price in sales)

# 使い方の例
if __name__ == "__main__":
    today = [
        ("りんご", 10, 120),
        ("バナナ", 5, 80),
        ("みかん", 12, 60),
    ]
    total = calculate_daily_sales(today)
    print(f"本日の売上合計: {total} 円")
```
